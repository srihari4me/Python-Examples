```python
ab_df.drop(['id','host_id','neighbourhood_group','neighbourhood','last_review','name','host_name'],axis=1, inplace=True)
```

```python
ab_df['reviews_per_month']=ab_df['reviews_per_month'].fillna(0)
ab_df['number_of_reviews']=ab_df['number_of_reviews'].fillna(0)
ab_df.isnull().sum()
```

```python
ab_df['room_type'].unique() 
```

```python
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
ab_df['room_type']= label_encoder.fit_transform(ab_df['room_type']) 
  
ab_df['room_type'].unique() 
```
